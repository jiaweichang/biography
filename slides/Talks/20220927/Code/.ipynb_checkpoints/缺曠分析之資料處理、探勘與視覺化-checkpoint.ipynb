```python
#重新命名成英文並存成一個 Dict 容器
new_col_name = {'學期':'Year_Semester', '班級':'Year_Class', '學號':'Student_ID', '科目名稱':'Subject_Name', 
                '開課單位':'Course_Unit', '課程時數':'Course_Hours', '缺課時數':'Absence_Hours'}
```

```python
#將欄位重新命名成英文
data = data.rename(columns=new_col_name)
```

```python
data
```

```python
data['Course_Total_Hours'] = data['Course_Hours']*18
```

```python
data['Absence_Ratio'] = data['Absence_Hours']/data['Course_Total_Hours']
```

```python
data
```

```python
data.dtypes #顯示出所有屬性欄位的資料型態
```

```python
data.dtypes.value_counts() #統計出資料中的各種資料型態
```

```python
data.info() #統計出所有欄位的值，可以發現 空值(nan) 的線索
```

```python
AbsenceStatistics_byCourse = (data
     .groupby(['Year_Semester','Year_Class','Subject_Name','Course_Unit', 'Course_Hours'])
     .agg({'Absence_Hours':['count','max','min','mean','median', 'std'],
           'Absence_Ratio':['count','max','min','mean','median', 'std']})
     .fillna(0)
)
```

```python
AbsenceStatistics_byCourse
```

```python
AbsenceStatistics_byCourse.head(10)
```

```python
AbsenceStatistics_byCourse[AbsenceStatistics_byCourse['Absence_Hours']['count']>9]
```

```python
AbsenceStatistics_byCourse[(AbsenceStatistics_byCourse['Absence_Ratio']['mean']>0.5) & 
                           (AbsenceStatistics_byCourse['Absence_Ratio']['count']>10)]
```

```python
AbsenceStatistics_10701 = data[data['Year_Semester']=="107-1"]
```

```python
AbsenceStatistics_10701
```

```python
AbsenceStatistics_by_10701_Year_Class = (AbsenceStatistics_10701
     .groupby(['Year_Class'])
     .agg({'Absence_Hours':['count','max','min','mean','median', 'std'],
           'Absence_Ratio':['count','max','min','mean','median', 'std']})
     .fillna(0)
)
```

```python
AbsenceStatistics_by_10701_Year_Class
```

```python
AbsenceStatistics_by_10701_Year_Class[(AbsenceStatistics_by_10701_Year_Class['Absence_Ratio']['mean']>0.5) & 
                                      (AbsenceStatistics_by_10701_Year_Class['Absence_Ratio']['count']>10)]
```

```python
AbsenceStatistics_by_10701_Course_Unit = (AbsenceStatistics_10701
     .groupby(['Course_Unit'])
     .agg({'Absence_Hours':['count','max','min','mean','median', 'std'],
           'Absence_Ratio':['count','max','min','mean','median', 'std']})
     .fillna(0)
)
```

```python
AbsenceStatistics_by_10701_Course_Unit[(AbsenceStatistics_by_10701_Course_Unit['Absence_Ratio']['mean']>0.5) 
                                       & (AbsenceStatistics_by_10701_Course_Unit['Absence_Ratio']['count']>10)]
```

```python
 #以 Dict 容器，建立臺中地區的郵遞區號列表
Post_Taichung = {400:"中區", 401:"東區", 402:"南區", 403:"西區", 404:"北區", 406:"北屯區", 407:"西屯區", 408:"南屯區", 411:"太平區", 412:"大里區", 413:"霧峰區", 414:"烏日區", 420:"豐原區", 421:"后里區", 422:"石岡區", 423:"東勢區",424:"和平區",426:"新社區",427:"潭子區",428:"大雅區",429:"神岡區",432:"大肚區",433:"沙鹿區",434:"龍井區",435:"梧棲區",436:"清水區",437:"大甲區",438:"外埔區",439:"大安區"}
```

```python
PostCode_Taichung = list(Post_Taichung.keys()) #將 Post_Taichung 的所有 Key 列出並轉成 list 容器存成 PostCode_Taichung
```

```python
PostCode_Taichung = list(map(str,PostCode_Taichung)) #將 PostCode_Taichung 的所有key 從 int 轉成 str 型態
```

```python
#指定選取通訊郵遞區號屬於臺中地區的學生資料
NewStudent_107_Taichung = NewStudent_107[NewStudent_107.Corresponding_PostCode.isin(PostCode_Taichung)] 
```

```python
NewStudent_107_Taichung
```

```python
#統計各個臺中地區郵遞區號的學生人數
NewStudent_107_Taichung.Corresponding_PostCode.value_counts() 
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
sns.set(font=['sans-serif'])
sns.set_style("whitegrid",{"font.sans-serif":['Microsoft JhengHei']})
```

```python
#利用通訊郵遞區號來找出 107學年度十大新生人數最多的臺中區域
fig, ax = plt.subplots(figsize=(10,12))
top_n = NewStudent_107_Taichung.Corresponding_PostCode.value_counts().index[:10]

sns.countplot(y='Corresponding_PostCode', 
              data = (NewStudent_107.assign(
                  Corresponding_PostCode=NewStudent_107.Corresponding_PostCode.where(NewStudent_107.Corresponding_PostCode.isin(top_n),'Other')
              )
                     )
             )
```

```python
top_n
```

```python
#統計來自各畢業高中職的新生人數
NewStudent_107.Graduated_School.value_counts()
```

```python
#利用畢業學校來找出 107學年度十大新生人數最多的生源學校，但有資料未統一的問題
fig, ax = plt.subplots(figsize=(10,12))
top_n_school = NewStudent_107_Taichung.Graduated_School.value_counts().index[:10]

sns.countplot(y='Graduated_School', 
              data = (NewStudent_107.assign(
                  Graduated_School = NewStudent_107_Taichung.Graduated_School.where(NewStudent_107_Taichung.Graduated_School.isin(top_n_school), 'Other')
              )
                     )
             )
```

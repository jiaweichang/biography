```python
from scipy import stats
```

```python
Admission_Method_1 = corr_between_RankAbsence[corr_between_RankAbsence['Admission_Method']=="甄選入學"]
```

```python
Admission_Method_2 = corr_between_RankAbsence[corr_between_RankAbsence['Admission_Method']=="聯合分發"]
```

```python
len(Admission_Method_1)
```

```python
len(Admission_Method_2)
```

```python
stats.levene(Admission_Method_1['Rank_10801'], Admission_Method_2['Rank_10801']) 
#Levene's test 同質性檢定，檢驗不顯著（p 值很大），這意味著方差具有同質性。
```

```python
# Shapiro-Wilk 正態性檢驗，p值很小，不滿足常態分布
stats.shapiro(Admission_Method_1['Rank_10801']) 
```

```python
# Shapiro-Wilk 正態性檢驗，p值很小，不滿足常態分布
stats.shapiro(Admission_Method_2['Rank_10801'])
```

```python
stats.ttest_ind(Admission_Method_1['Rank_10801'], Admission_Method_2['Rank_10801']) #獨立樣本 T 檢定，p值顯示無顯著差異
```
